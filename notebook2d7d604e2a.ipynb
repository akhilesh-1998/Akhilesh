```python
#concate the above Yearly converted salary
Yearly=pd.concat([weekly,monthly,yearly])
```

```python
#Compaerion b/w databse use this year and will be required in next year.
# Language worked with 
data = pd.DataFrame(dataP['DatabaseWorkedWith'].dropna().str.split(';').tolist()).stack()
data=data.value_counts().sort_values(ascending=False).head(10).plot(kind='bar')
# Ploting the graph
plt.title('DatabaseWorkedWith')
plt.tight_layout()
plt.show()

# Demanidng data base for this year is MYSQL
```
